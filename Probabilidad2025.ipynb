## 5Ô∏è‚É£ **Reglas de Probabilidad** üìä

### üìå **Complemento**

$$

$$
Ejemplo: Si (P(lluvia) = 0.3), entonces (P(no\ lluvia) = 0.7).

---

### üìå **Uni√≥n de eventos (adici√≥n)**

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

‚úÖ **Ejemplo resuelto:**
Lanzar un dado.

* A = ‚Äún√∫mero par‚Äù ‚Üí (P(A) = 3/6).
* B = ‚Äún√∫mero mayor que 4‚Äù ‚Üí (P(B) = 2/6).
* Intersecci√≥n: {6} ‚Üí (P(A \cap B) = 1/6).

$$
P(A \cup B) = \frac{3}{6} + \frac{2}{6} - \frac{1}{6} = \frac{4}{6} = \frac{2}{3}
$$

---

### üìå **Intersecci√≥n de eventos (independientes)**

$$
P(A \cap B) = P(A) \cdot P(B)
$$

‚úÖ **Ejemplo resuelto:**
Lanzar dos monedas.

* A = ‚Äúprimera cara‚Äù ‚Üí (P(A) = 1/2).
* B = ‚Äúsegunda cara‚Äù ‚Üí (P(B) = 1/2).

$$
P(A \cap B) = (1/2)(1/2) = 1/4
$$

## 8Ô∏è‚É£ **Definiciones de Probabilidad** üîé

Existen **tres enfoques principales**: **Cl√°sica (Laplace), Frecuencial y Axiom√°tica**.

---

### üìê **1. Probabilidad Cl√°sica (Laplace)**

üëâ Se aplica cuando **todos los resultados son equiprobables**.

$$
P(E) = \frac{\text{Casos favorables}}{\text{Casos posibles}}
$$

üé≤ **Ejemplo:**
Lanzar un dado y calcular la probabilidad de obtener un n√∫mero par.
$$
P(E) = \frac{3}{6} = \tfrac{1}{2}
$$

---

### üìä **2. Probabilidad Frecuencial**

üëâ La probabilidad se entiende como el **l√≠mite de la frecuencia relativa** cuando el n√∫mero de repeticiones del experimento tiende a infinito.

$$
P(E) = \lim_{n \to \infty} \frac{f(E)}{n}
$$

donde (f(E)) = n√∫mero de veces que ocurre el evento E, y (n) = n√∫mero total de repeticiones.

‚öΩ **Ejemplo real:**
Un equipo de f√∫tbol gana 45 de 100 partidos jugados.
$$
P(Ganar) \approx \frac{45}{100} = 0.45
$$

üëâ Si el equipo sigue jugando m√°s partidos, la frecuencia relativa tender√° al valor real de la probabilidad.

‚úÖ **Ejemplo visual:**
üìà En una simulaci√≥n de lanzar una moneda 1.000 veces:

* 503 veces ‚Üí cara
* 497 veces ‚Üí sello
  $$
  P(\text{cara}) \approx \frac{503}{1000} = 0.503
  $$
  Muy cercano al valor te√≥rico (0.5).

---

### üìö **3. Probabilidad Axiom√°tica (Kolmog√≥rov)**

üëâ Es el enfoque **m√°s riguroso y general**, basado en tres axiomas:

1Ô∏è‚É£ **No negatividad:**
$$
P(E) \geq 0
$$

2Ô∏è‚É£ **Normalizaci√≥n:**
$$
P(S) = 1
$$

3Ô∏è‚É£ **Aditividad (eventos mutuamente excluyentes):**
$$
P(A \cup B) = P(A) + P(B), \quad \text{si } A \cap B = \varnothing
$$

---

üîπ **Ejemplo:**
En un dado justo:

* Evento A = {1,2}
* Evento B = {3}
* Evento C = {4,5,6}

Se cumplen:

* $P(S) = 1$
* $P(A) = 2/6 = 1/3$
* $P(B) = 1/6$
* $P(C) = 3/6 = 1/2$
* $P(A) + P(B) + P(C) = 1$ ‚úÖ

# üìò **Probabilidad Condicional, Regla de la Probabilidad Total y Teorema de Bayes**

---

## üéØ **1. Probabilidad Condicional**

La **probabilidad condicional** mide la probabilidad de que ocurra un evento (A), dado que ya ocurri√≥ otro evento (B).

üìå **Definici√≥n:**

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}, \quad P(B) > 0
$$

üëâ Esto significa: *la probabilidad de (A) bajo la condici√≥n de que (B) haya ocurrido*.

üîπ **Ejemplo (con √≠cono üé≤):**
Se lanza un dado.

* $A = {\text{salir n√∫mero par}}$
* $B = {\text{salir mayor que 3}}$

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$$

* $A = {2,4,6}$
* $B = {4,5,6}$
* $A \cap B = {4,6}$

$$
P(A|B) = \frac{2/6}{3/6} = \frac{2}{3}
$$

üéØ **Interpretaci√≥n:** Si sabemos que sali√≥ un n√∫mero mayor que 3, la probabilidad de que sea par es **$( \tfrac{2}{3} )$**.

---

## üåê **2. Regla de la Probabilidad Total**

La **probabilidad total** se usa cuando un evento (A) puede descomponerse en funci√≥n de varios escenarios mutuamente excluyentes (B_1, B_2, ..., B_n).

üìå **Definici√≥n:**

$$
P(A) = \sum_{i=1}^{n} P(A|B_i) \cdot P(B_i)
$$

üîπ **Ejemplo (con √≠cono üè•):**
Un hospital hace pruebas de detecci√≥n:

* El **60%** de los pacientes son hombres ((B_1)) y el **40%** mujeres ((B_2)).
* La probabilidad de dar positivo en la prueba es:

  * $P(Positivo|Hombre) = 0.1$
  * $P(Positivo|Mujer) = 0.2$

¬øCu√°l es la probabilidad de que un paciente al azar d√© positivo?

$$
P(Positivo) = P(Positivo|Hombre)\cdot P(Hombre) + P(Positivo|Mujer)\cdot P(Mujer)
$$

$$
P(Positivo) = (0.1)(0.6) + (0.2)(0.4) = 0.06 + 0.08 = 0.14
$$

üëâ La probabilidad de dar positivo en general es **14%**.

---

## üé≤ **3. Teorema de Bayes**

El **teorema de Bayes** permite invertir probabilidades condicionales: calcular la probabilidad de una **causa** dado un **efecto**.

üìå **Definici√≥n:**

$$
P(B_j|A) = \frac{P(A|B_j) \cdot P(B_j)}{\sum_{i=1}^{n} P(A|B_i) \cdot P(B_i)}
$$

üîπ **Ejemplo (con √≠cono üß™):**
Un laboratorio tiene **2 m√°quinas** que producen reactivos:

* M√°quina 1 produce el 70% ((B_1)) con 2% defectuosos.
* M√°quina 2 produce el 30% ((B_2)) con 5% defectuosos.

Si un reactivo se toma al azar y resulta **defectuoso**, ¬øcu√°l es la probabilidad de que venga de la m√°quina 2?

* $P(Def|B_1) = 0.02$, $P(B_1) = 0.7$
* $P(Def|B_2) = 0.05$, $P(B_2) = 0.3$

$$
P(B_2|Def) = \frac{P(Def|B_2)\cdot P(B_2)}{P(Def|B_1)P(B_1) + P(Def|B_2)P(B_2)}
$$

$$
P(B_2|Def) = \frac{0.05 \cdot 0.3}{0.02\cdot0.7 + 0.05\cdot0.3} = \frac{0.015}{0.014+0.015} = \frac{0.015}{0.029} \approx 0.517
$$

üëâ Hay un **51.7%** de probabilidad de que el producto defectuoso venga de la m√°quina 2.

## üé≤ **3. Teorema de Bayes**

$$
