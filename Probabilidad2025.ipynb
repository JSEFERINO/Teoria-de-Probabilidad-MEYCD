# 📘 **Guía de Clase – Probabilidad Básica**

Comprender los fundamentos de la probabilidad desde la definición de experimento aleatorio hasta las principales reglas de probabilidad, mediante ejemplos prácticos y ejercicios.

## 1️⃣ **Experimento Aleatorio** 🎲

👉 Un **experimento aleatorio** es aquel cuyo resultado **no puede predecirse con certeza**, aunque sabemos qué resultados son posibles.

**Ejemplos:**

* 🎲 Lanzar un dado → los resultados posibles son {1,2,3,4,5,6}.
* 🪙 Lanzar una moneda → {cara, sello}.
* 🃏 Extraer una carta de una baraja → 52 resultados posibles.

---

## 2️⃣ **Espacio Muestral y Eventos** 📋

* **Espacio muestral (S):** conjunto de todos los posibles resultados.
* **Evento (E):** subconjunto del espacio muestral.

**Ejemplo (lanzar un dado):**

* Espacio muestral: ( S = {1,2,3,4,5,6} ).
* Evento A = “sacar un número par” → ( A = {2,4,6} ).
* Evento B = “sacar un número mayor que 4” → ( B = {5,6} ).

✅ **Ejercicio resuelto:**
¿Cuál es el número de elementos del espacio muestral al lanzar dos monedas?
$$
S = {CC, CS, SC, SS} \quad \Rightarrow \quad |S| = 4
$$

---

## 3️⃣ **Probabilidad según Laplace** 📐

Cuando los resultados son **igualmente probables**:
$$
P(E) = \frac{\text{Casos favorables}}{\text{Casos posibles}}
$$

**Ejemplo:**

* Lanzar un dado.
* Evento: “obtener un número par” = {2,4,6}.
  $$
  P(E) = \frac{3}{6} = \frac{1}{2}
  $$

✅ **Ejercicio resuelto:**
¿Qué probabilidad hay de obtener “corazón” al extraer una carta de una baraja?
$$
P(E) = \frac{13}{52} = \frac{1}{4}
$$

---

## 4️⃣ **Tipos de Eventos** 🔍

* **Evento seguro:** ocurre siempre → lanzar un dado y obtener un número entre 1 y 6.
* **Evento imposible:** nunca ocurre → lanzar un dado y obtener un 7.
* **Eventos mutuamente excluyentes:** no pueden suceder juntos → obtener cara e inmediatamente sello en un mismo lanzamiento de moneda.

---

## 5️⃣ **Reglas de Probabilidad** 📊

### 📌 **Complemento**

$$
P(A^c) = 1 - P(A)
$$
Ejemplo: Si (P(lluvia) = 0.3), entonces (P(no\ lluvia) = 0.7).

---

### 📌 **Unión de eventos (adición)**

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

✅ **Ejemplo resuelto:**
Lanzar un dado.

* A = “número par” → (P(A) = 3/6).
* B = “número mayor que 4” → (P(B) = 2/6).
* Intersección: {6} → (P(A \cap B) = 1/6).

$$
P(A \cup B) = \frac{3}{6} + \frac{2}{6} - \frac{1}{6} = \frac{4}{6} = \frac{2}{3}
$$

---

### 📌 **Intersección de eventos (independientes)**

$$
P(A \cap B) = P(A) \cdot P(B)
$$

✅ **Ejemplo resuelto:**
Lanzar dos monedas.

* A = “primera cara” → (P(A) = 1/2).
* B = “segunda cara” → (P(B) = 1/2).

$$
P(A \cap B) = (1/2)(1/2) = 1/4
$$

---

## 8️⃣ **Definiciones de Probabilidad** 🔎

Existen **tres enfoques principales**: **Clásica (Laplace), Frecuencial y Axiomática**.

---

### 📐 **1. Probabilidad Clásica (Laplace)**

👉 Se aplica cuando **todos los resultados son equiprobables**.

$$
P(E) = \frac{\text{Casos favorables}}{\text{Casos posibles}}
$$

🎲 **Ejemplo:**
Lanzar un dado y calcular la probabilidad de obtener un número par.
$$
P(E) = \frac{3}{6} = \tfrac{1}{2}
$$

---

### 📊 **2. Probabilidad Frecuencial**

👉 La probabilidad se entiende como el **límite de la frecuencia relativa** cuando el número de repeticiones del experimento tiende a infinito.

$$
P(E) = \lim_{n \to \infty} \frac{f(E)}{n}
$$

donde (f(E)) = número de veces que ocurre el evento E, y (n) = número total de repeticiones.

⚽ **Ejemplo real:**
Un equipo de fútbol gana 45 de 100 partidos jugados.
$$
P(Ganar) \approx \frac{45}{100} = 0.45
$$

👉 Si el equipo sigue jugando más partidos, la frecuencia relativa tenderá al valor real de la probabilidad.

✅ **Ejemplo visual:**
📈 En una simulación de lanzar una moneda 1.000 veces:

* 503 veces → cara
* 497 veces → sello
  $$
  P(\text{cara}) \approx \frac{503}{1000} = 0.503
  $$
  Muy cercano al valor teórico (0.5).

---

### 📚 **3. Probabilidad Axiomática (Kolmogórov)**

👉 Es el enfoque **más riguroso y general**, basado en tres axiomas:

1️⃣ **No negatividad:**
$$
P(E) \geq 0
$$

2️⃣ **Normalización:**
$$
P(S) = 1
$$

3️⃣ **Aditividad (eventos mutuamente excluyentes):**
$$
P(A \cup B) = P(A) + P(B), \quad \text{si } A \cap B = \varnothing
$$

---

🔹 **Ejemplo:**
En un dado justo:

* Evento A = {1,2}
* Evento B = {3}
* Evento C = {4,5,6}

Se cumplen:

* $P(S) = 1$
* $P(A) = 2/6 = 1/3$
* $P(B) = 1/6$
* $P(C) = 3/6 = 1/2$
* $P(A) + P(B) + P(C) = 1$ ✅

---

## 9️⃣ **Ejercicios Resueltos** ✍️

   Un jugador de baloncesto encesta 72 tiros de 120 intentos. Estime la probabilidad de encestar.

   $$

2. **Axiomas** 🎲
   En un dado justo:

* A = “número par” → {2,4,6}
* B = “número impar” → {1,3,5}

Se cumple que:
$$
P(A) = \tfrac{3}{6} = \tfrac{1}{2}, \quad P(B) = \tfrac{3}{6} = \tfrac{1}{2}
$$
y
$$
P(A \cup B) = P(A) + P(B) = 1
$$

3. **Frecuencial vs. Clásica** 🪙
   Al lanzar una moneda 200 veces, se obtuvo 110 caras y 90 sellos.

* Definición clásica: (P(cara) = 0.5).
* Definición frecuencial: (P(cara) \approx \tfrac{110}{200} = 0.55).

👉 Diferencia: la clásica es **teórica**, la frecuencial es **empírica**.

# 📘 **Probabilidad Condicional, Regla de la Probabilidad Total y Teorema de Bayes**

---

## 🎯 **1. Probabilidad Condicional**

La **probabilidad condicional** mide la probabilidad de que ocurra un evento (A), dado que ya ocurrió otro evento (B).

📌 **Definición:**

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}, \quad P(B) > 0
$$

👉 Esto significa: *la probabilidad de (A) bajo la condición de que (B) haya ocurrido*.

🔹 **Ejemplo (con ícono 🎲):**
Se lanza un dado.

* $A = {\text{salir número par}}$
* $B = {\text{salir mayor que 3}}$

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$$

* $A = {2,4,6}$
* $B = {4,5,6}$
* $A \cap B = {4,6}$

$$
P(A|B) = \frac{2/6}{3/6} = \frac{2}{3}
$$

🎯 **Interpretación:** Si sabemos que salió un número mayor que 3, la probabilidad de que sea par es **$( \tfrac{2}{3} )$**.

---

## 🌐 **2. Regla de la Probabilidad Total**

La **probabilidad total** se usa cuando un evento (A) puede descomponerse en función de varios escenarios mutuamente excluyentes (B_1, B_2, ..., B_n).

📌 **Definición:**

$$
P(A) = \sum_{i=1}^{n} P(A|B_i) \cdot P(B_i)
$$

🔹 **Ejemplo (con ícono 🏥):**
Un hospital hace pruebas de detección:

* El **60%** de los pacientes son hombres ((B_1)) y el **40%** mujeres ((B_2)).
* La probabilidad de dar positivo en la prueba es:

  * $P(Positivo|Hombre) = 0.1$
  * $P(Positivo|Mujer) = 0.2$

¿Cuál es la probabilidad de que un paciente al azar dé positivo?

$$
P(Positivo) = P(Positivo|Hombre)\cdot P(Hombre) + P(Positivo|Mujer)\cdot P(Mujer)
$$

$$
P(Positivo) = (0.1)(0.6) + (0.2)(0.4) = 0.06 + 0.08 = 0.14
$$

👉 La probabilidad de dar positivo en general es **14%**.

---

## 🎲 **3. Teorema de Bayes**

El **teorema de Bayes** permite invertir probabilidades condicionales: calcular la probabilidad de una **causa** dado un **efecto**.

📌 **Definición:**

$$
P(B_j|A) = \frac{P(A|B_j) \cdot P(B_j)}{\sum_{i=1}^{n} P(A|B_i) \cdot P(B_i)}
$$

🔹 **Ejemplo (con ícono 🧪):**
Un laboratorio tiene **2 máquinas** que producen reactivos:

* Máquina 1 produce el 70% ((B_1)) con 2% defectuosos.
* Máquina 2 produce el 30% ((B_2)) con 5% defectuosos.

Si un reactivo se toma al azar y resulta **defectuoso**, ¿cuál es la probabilidad de que venga de la máquina 2?

* $P(Def|B_1) = 0.02$, $P(B_1) = 0.7$
* $P(Def|B_2) = 0.05$, $P(B_2) = 0.3$

$$
P(B_2|Def) = \frac{P(Def|B_2)\cdot P(B_2)}{P(Def|B_1)P(B_1) + P(Def|B_2)P(B_2)}
$$

$$
P(B_2|Def) = \frac{0.05 \cdot 0.3}{0.02\cdot0.7 + 0.05\cdot0.3} = \frac{0.015}{0.014+0.015} = \frac{0.015}{0.029} \approx 0.517
$$

👉 Hay un **51.7%** de probabilidad de que el producto defectuoso venga de la máquina 2.

# 📘 **Guía Extendida de Ejemplos**

---

## 🎯 **1. Probabilidad Condicional**

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$$

### Ejemplos:

1. 🎲 **Dados**: Se lanza un dado.

   * $A={\text{número par}}, B={\text{número > 4}}$.
   * $A \cap B = {6}$.
   * $P(A|B) = \tfrac{1/6}{2/6} = \tfrac{1}{2}$.

2. 🃏 **Cartas**: De una baraja de 52 cartas, ¿cuál es la probabilidad de que salga un rey si sabemos que salió una figura (J,Q,K)?

   * $P(K|Figura) = \tfrac{4/52}{12/52} = \tfrac{1}{3}$.

3. 🏫 **Estudiantes**: En una clase, el 40% son hombres, y el 20% del total son hombres con gafas.

   * $P(Gafas|Hombre) = \tfrac{0.20}{0.40} = 0.5$.

4. ⚽ **Deportes**: De 100 partidos, 60 fueron en casa y 30 de esos se ganaron.

   * $P(Ganar|Casa) = \tfrac{30}{60} = 0.5$.

5. 💊 **Medicina**: El 10% de la población está enferma. Entre los enfermos, el 80% tiene fiebre.

   * $P(Fiebre|Enfermo) = 0.8$.

## 🌐 **2. Regla de la Probabilidad Total**

$$
P(A) = \sum_{i=1}^{n} P(A|B_i)\cdot P(B_i)
$$

### Ejemplos:

1. 🏥 **Hospital**:

   * $P(B_1)=0.7, ; P(B_2)=0.3$.
   * $P(Curar|A)=0.9, ; P(Curar|B)=0.8$.
   * $P(Curar) = 0.9\cdot0.7 + 0.8\cdot0.3 = 0.87$.

2. 🚗 **Seguro**:

   * $P(Nuevo)=0.6, ; P(Usado)=0.4$.
   * $P(Acc|Nuevo)=0.05, ; P(Acc|Usado)=0.15$.
   * $P(Accidente) = 0.05\cdot0.6 + 0.15\cdot0.4 = 0.09$.

3. 🏫 **Escuelas**:

   * A: $P=0.5, P(Aprobar|A)=0.8$
   * B: $P=0.3, P(Aprobar|B)=0.7$
   * C: $P=0.2, P(Aprobar|C)=0.6$
   * $P(Aprobar) = 0.73$.

4. 🎲 **Dados sesgados**:

   * $P(A)=0.4, ; P(6|A)=0.2$
   * $P(B)=0.6, ; P(6|B)=0.1$
   * $P(6) = 0.14$.

5. 🧪 **Test médico**:

   * $P(Enf)=0.05, P(Pos|Enf)=0.9$
   * $P(Sano)=0.95, P(Pos|Sano)=0.1$
   * $P(Pos) = 0.14$.

---

## 🎲 **3. Teorema de Bayes**

$$
P(B_j|A) = \frac{P(A|B_j) \cdot P(B_j)}{\sum_{i=1}^{n} P(A|B_i)\cdot P(B_i)}
$$

### Ejemplos:

1. 🏥 **Enfermedad**:

   * $P(Enf)=0.05, P(Pos|Enf)=0.9, P(Pos|Sano)=0.1$.
   * $P(Enf|Pos) = \dfrac{0.9\cdot0.05}{0.9\cdot0.05+0.1\cdot0.95} = \dfrac{0.045}{0.14} \approx 0.321$.

2. 🚗 **Seguro**:

   * $P(Lujo)=0.2, P(Acc|Lujo)=0.1$
   * $P(Comun)=0.8, P(Acc|Comun)=0.05$
   * $P(Lujo|Acc) = \dfrac{0.1\cdot0.2}{0.1\cdot0.2+0.05\cdot0.8} = \tfrac{1}{3}$.

3. 📚 **Examen**:

   * A: $0.5, P(Aprobar|A)=0.9$
   * B: $0.3, P(Aprobar|B)=0.7$
   * C: $0.2, P(Aprobar|C)=0.6$
   * $P(A|Aprobar)=\dfrac{0.9\cdot0.5}{0.9\cdot0.5+0.7\cdot0.3+0.6\cdot0.2} \approx 0.577$.

4. ⚽ **Deportes**:

   * $P(Fuerte)=0.4, P(Ganar|Fuerte)=0.7$
   * $P(Débil)=0.6, P(Ganar|Débil)=0.3$
   * $P(Fuerte|Ganar) = \dfrac{0.28}{0.46} \approx 0.609$.

5. 🧪 **Calidad industrial**:

   * M1: $0.5, P(Def|M1)=0.02$
   * M2: $0.3, P(Def|M2)=0.04$
   * M3: $0.2, P(Def|M3)=0.05$
   * $P(M3|Def) = \dfrac{0.05\cdot0.2}{0.01+0.012+0.01} \approx 0.312$.
